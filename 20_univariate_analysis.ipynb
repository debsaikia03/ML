```python
sns.countplot(df, x='Survived')
df['Survived'].value_counts()
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df, x='Pclass')
```

```python
sns.countplot(df, x='Sex')
```

```python
df['Survived'].value_counts().plot(kind='pie', autopct='%.2f')
```
