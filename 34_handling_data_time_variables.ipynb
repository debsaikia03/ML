```python
#working with dates

# Converting to datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

```python
#1. extract year

date['date_year'] = date['date'].dt.year

date.sample(5)
```

```python
#2. extract month

date['date_month_no'] = date['date'].dt.month

date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()

date.head()
```

```python
# extract days

date['date_day'] = date['date'].dt.day

date.head()
```

```python
# day of week
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

```python
# day of week - name

date['date_dow_name'] = date['date'].dt.day_name()

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#extract week of the year

# Extract week of the year using isocalendar().week
date['date_week'] = date['date'].dt.isocalendar().week

# Display the first few rows
date.head()
```

```python
#extract quarter

# Extract quarter of the year
date['quarter'] = date['date'].dt.quarter
date.head()
```

```python
#extract semester

date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

```python
#months passed

from datetime import datetime

# Ensure date['date'] is a datetime object
date['date'] = pd.to_datetime(date['date'])

# Get today's date
today = pd.Timestamp(datetime.now())

# Calculate calendar-based months passed
(today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month)
```
