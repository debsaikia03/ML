```python
#dropping missing data using dropna() 
```

```python
import numpy as np
import pandas as pd

data= pd.read_csv('titanic.csv')
data
```

```python
data.shape
```

```python
data.dropna()
```

```python
data.dropna().shape
```

```python
data.dropna(how='all').shape #remove entry where a row is all NaN
```

```python
data.dropna(axis=1) #remove all columns where it contains even 1 missing value
```

```python
data.dropna(axis=1).shape #12 -> 7
```

```python
data.dropna(axis=1, how='all').shape ##remove that column where it contains all missing values in that column
```

```python
data.dropna(subset=['Cabin']).shape
```

```python
data.dropna(subset=['Cabin','Embarked']).shape
```

```python
#fillna() -> filling missing values 
```

```python
data
```

```python
data.fillna(0)
```

```python
data['Cabin'].fillna('Not Specified', inplace=True) #apply individuak column
data
```

```python
data['Age'] = data['Age'].bfill()
data
```

```python
data['Age'] = data['Age'].ffill()
data
```
